### **1) INVESTIGACIÓN DE OPERACIONES** 
Modelo de Optimización de Recursos

```python
import numpy as np
from scipy.optimize import linprog
```

```python
# Ruta archivos CSV
personal = 'https://raw.githubusercontent.com/Victor4dan/Residuos-CDMX-URC/main/salarios%20promedio.csv'
personal = pd.read_csv(personal, encoding='latin1')
```

**Paso 1: Definir Variables**
Vamos a definir las variables de decisión de la siguiente manera:

- 𝑥1: Número de empleados base o sindicalizados
- 𝑥2 :Número de empleados eventuales
- x3: Número de empleados en otras categorías

**Paso 2: Definir la Función Objetivo**
La función objetivo es minimizar el costo total del personal. Supongamos que los costos asociados con cada tipo de contrato son:
- C1: Costo por empleado base o sindicalizado
- C2: Costo por empleado eventual
- C3: Costo por empleado en otras categorías

Entonces, la función objetivo es: 

 **Minimizar Z= (C1 * x1) + (C2 * x2 ) + (C3 * x3)**
​


```python
# Obtener el máximo y mínimo de las columnas especificadas para crear las restricciones
max_base_sindicalizado = personal['Base o sindicalizado'].max()
min_base_sindicalizado = personal['Base o sindicalizado'].min()

max_eventual = personal['Eventual'].max()
min_eventual = personal['Eventual'].min()

max_otros = personal['Otros '].max()
min_otros = personal['Otros '].min()
```

```python
# Imprimir los resultados
print("Máximo de 'Base o sindicalizado':", max_base_sindicalizado)
print("Mínimo de 'Base o sindicalizado':", min_base_sindicalizado)

print("Máximo de 'Eventual':", max_eventual)
print("Mínimo de 'Eventual':", min_eventual)

print("Máximo de 'Otros':", max_otros)
print("Mínimo de 'Otros':", min_otros)
```

**Paso 3: Definir las Restricciones**

Las restricciones están basadas en los máximos y mínimos de cada tipo de contrato:


- 1.Restricción de empleados base o sindicalizados:

**$919,664 ≤ x1 ≤ $8,774,960**

- 2.Restricción de empleados eventuales:

**1 ≤ x 2 ≤ $5,040,080**

- 3.Restricción de empleados en otras categorías:

**1 ≤ x 3≤ $7,168,560**

**Paso 4: Formulación del Problema**

El problema de optimización se puede formular de la siguiente manera:

Minimizar Z=( C1* x1 )+( C2 * x2 )+(C3 * x3)
​
sujeto a:

- 1) $919,664 ≤ x1 ≤ $8,774,960
- 2) $1 ≤ x2 ≤ $5,040,080
- 3) $1≤ x3 ≤ $7,168,560

```python
# Definir la función objetivo: minimizar el total de empleados
c = np.array([1, 1, 1])
```

```python
# Definir las restricciones de desigualdad
A_ub = [
    [-1, 0, 0],  # x1 >= 919664 -> -x1 <= -919664
    [1, 0, 0],   # x1 <= 8774960
    [0, -1, 0],  # x2 >= 1 -> -x2 <= -1
    [0, 1, 0],   # x2 <= 5040080
    [0, 0, -1],  # x3 >= 1 -> -x3 <= -1
    [0, 0, 1]    # x3 <= 7168560
]

# Términos independientes para las restricciones de desigualdad
b_ub = [
    -919664,     # x1 >= 919664
    8774960,     # x1 <= 8774960
    -1,          # x2 >= 1
    5040080,     # x2 <= 5040080
    -1,          # x3 >= 1
    7168560      # x3 <= 7168560
]

# Limitar las variables a valores no negativos y dentro de los rangos especificados
x_bounds = [(919664, 8774960), (1, 5040080), (1, 7168560)]

# Resolver el problema de minimización
result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=x_bounds, method='simplex')

# Resultados
if result.success:
    print("Solución óptima encontrada:")
    print("Número de empleados base o sindicalizados:", result.x[0])
    print("Número de empleados eventuales:", result.x[1])
    print("Número de empleados en otras categorías:", result.x[2])
    print("Cantidad total mínima de empleados:", result.fun)
else:
    print("No se encontró una solución óptima.")
```
